```python
for prop in root.findall("property"):
    prop_name = prop.get("name")
    prop_type = prop.get("type")
    prop_key = prop.get("key", None) or prop_name
    if prop_type == "percent":
        print(f"PercentageProp('{prop_name}', '{prop_key}', '{prop_name}'),")
    elif prop_type == "int":
        print(f"IntegerProp('{prop_name}', '{prop_key}', '{prop_name}'),")
    elif prop_type == "bool":
        print(f"BooleanProp('{prop_name}', '{prop_key}', '{prop_name}'),")
```
